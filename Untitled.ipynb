```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error,mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
import catboost as cbt
from catboost import CatBoostRegressor, Pool, cv
import lightgbm as lgb
from lightgbm import LGBMRegressor
import numpy as np
import os
import panel as pn
from io import BytesIO
```

```python
#file_input.param.watch(top, options)
pn.extension()

#options = ['Bundles_19.csv',
#'Bundles_79_new.csv',
#'Bundles_47_new.csv',
#'Bundles_26-2.csv',
#'Bundles_13_new.csv',
#'Bundles_864.csv',
#'Bundles_858.csv',
#'Bundles_857.csv',
#'Bundles_855.csv',
#'Bundles_1.csv']
#file_input = pn.widgets.FileInput(sizing_mode='stretch_width')
file_input = pn.widgets.Select(name='', options=['Bundles_1.csv',
                                                     'Bundles_19.csv',
                                                     'Bundles_858.csv',
                                                     'Bundles_13_new.csv',
                                                     'Bundles_47_new.csv',
                                                     'Bundles_864.csv',
                                                     'Bundles_79_new.csv',
                                                     'Bundles_26-2.csv',
                                                     'Bundles_857.csv',
                                                     'Bundles_855.csv',
                                                     'Bundles_1.csv']) 

experience = pn.widgets.Select(name='', options={'  ': 4.0, ' ': 0.0, ' 1   3 ': 1.0, ' 3  6 ': 2.0, ' 6 ': 3.0})
button = pn.widgets.Button(name=' ')

#sidebar = pn.layout.WidgetBox(
#    pn.pane.Markdown("#  "),
#    file_input,
#    experience,
#    button,
#    pn.pane.Markdown("#### Carbon"),
#    #sizing_mode='stretch_width'
#).servable(area='sidebar')
#print(file_input)

#@pn.depends(file_input, watch=True)
def prep(file_input):
    #r'C:\Users\diana\Documents\ '
    if file_input is not None:
        df = pd.read_csv('/users/diana/documents/ /' + file_input, sep = ',').drop_duplicates()
#       df = pd.read_csv(BytesIO(file_input.value), sep = ',').drop_duplicates()

    null = df['name'].isnull()
    #     
    df = df[~null]

    pd.options.display.max_columns = None
    zero_cols = df.columns[(df == 0).all()]

    #   
    df.drop(zero_cols, axis=1, inplace=True)

    df.dropna(subset=['salary_from_rub', 'salary_to_rub'], how='all', inplace=True)
    df['salary'] = (df['salary_from_rub'] + df['salary_to_rub']) / 2
    df.loc[df['salary'].isnull(), 'salary'] = df.loc[df['salary'].isnull(), ['salary_from_rub', 'salary_to_rub']].apply(lambda x: x.dropna().iloc[0], axis=1)

    q_min = df.salary.quantile(0.03)
    q_max = df.salary.quantile(0.97)
    df = df[(df['salary'] > q_min) & (df['salary'] < q_max)]
    return df

#@pn.depends(file_input, experience)
def top(file_input, experience):
    df = prep(file_input.value)
    if experience.value != 4.0:
        df = df.loc[df['experience_id'] == experience.value]
    # 
    end = df.columns.get_loc('source_index')
    counts = df.iloc[:, df.columns.get_loc('id')+1:end].sum()
    topval = counts.sort_values(ascending=False)[:25]
    fig, ax = plt.subplots(1, 1, figsize=(7,8))
    fig.tight_layout(w_pad=5)
    sns.barplot(ax = ax, x=topval.values, y=topval.index, palette=sns.color_palette('magma', 25), alpha=0.7)
    ax.set_title(label=' ', loc='left', size=14, pad=15)
    ax.set_xlabel(' ')
    ax.set_ylabel('')
    return(ax.figure)
    #pn.pane.Matplotlib(plt)
    
def model():
    df = prep(file_input.value)
    skills = df.columns[df.columns.get_loc('id')+1:df.columns.get_loc('source_index')].tolist()
    X = df.iloc[:,df.columns.get_loc('id')+1:df.columns.get_loc('source_index')]
    y = df['salary']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    pools = {}
    pools['train'] = cbt.Pool(
        data=X_train,
        label=y_train
    )
    pools['test'] = cbt.Pool(
        data=X_test,
        label=y_test
    )
    model = CatBoostRegressor(iterations = 1000, eval_metric = 'MAE')
    model.fit(
        pools['train'],
        eval_set=pools['test'],
        use_best_model=True,
        #plot=True,
        verbose=500,
        early_stopping_rounds=50,
    )
    model_cat = pd.DataFrame([*zip(skills, model.feature_importances_[:len(skills)])], columns = ['features', 'values'])
    model_cat.sort_values(by='values', ascending=False, inplace=True)
    fig, ax = plt.subplots(1, 1, figsize=(7,8))
    sns.barplot(data=model_cat.head(25), x='values', y='features', palette=sns.color_palette('magma', 25), 
                orient='h', alpha=0.7, ax=ax)
    ax.set_title(label=f'     ', loc='left', size=14, pad=15)
    ax.set_xlabel(' . .')
    ax.set_ylabel('')
    return(ax.figure)
    

graf = top(file_input, experience)    
    
def btn_clc(val):
    print(file_input)
    print(experience)
    graf = top(file_input, experience)
    template.main[0] = graf
    print(file_input)
    print(experience)
    
button.on_click(btn_clc)

#file_input.param.watch(top, 'value')
#experience.param.watch(top, 'value')
#pn.bind(top, file_input, experience, Watch=True)

#template = pn.template.FastListTemplate(
#    title='  ',
#    sidebar,
#    sidebar=[pn.pane.Markdown("#  "),
#             file_input,
#             experience,
#             button
#             #pn.pane.Markdown("#### "), 
#             #pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),  
#    ],
#    main=[pn.Row(pn.Column(prep(file_input, experience), margin=(0,25))), 
#                co2_table.panel(width=500), 
#        pn.Row(pn.Column())], 
#                pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
#    accent_base_color="#97dae6",
#    header_background="#88d8b0",
#)
#main=pn.Column(pn.Row(pn.Column(top(file_input, experience), margin=(0,25))), 
##                co2_table.panel(width=500), 
#        pn.Row(pn.Column())).servable(title='  ')
#pn.Row(sidebar, main).show()
#print(file_input)
template = pn.template.FastListTemplate(
    title='  ', 
    sidebar=[pn.pane.Markdown("#  "),
             pn.pane.Markdown("#### ,     Panel,       ,    CSV ."),
             file_input,
             experience,
             button
    ],
#    main=[pn.Row(pn.Column(file_input)), 
    main=[pn.Row(pn.Column(graf, height=400, width=1000)), 
#    main=[pn.Row(pn.Column(top(file_input, experience))), 
#                 co2_table.panel(width=500)), 
          pn.Row(model, height=400, width=800) 
                 ],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
template.show()
template.servable();
```
